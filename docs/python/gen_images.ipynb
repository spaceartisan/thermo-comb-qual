```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters for the circle
r = 1  # Radius of the circle
buffer = 1.5  # Buffer to shift the circle into the top-right quadrant
theta = np.linspace(0, 2 * np.pi, 100)  # Angle from 0 to 2pi

# Parametric equations with offset to move the circle to the top-right quadrant
x = r * np.cos(theta) + r + buffer
y = r * np.sin(theta) + r + buffer

# Plot the circle
plt.figure()
plt.plot(x, y)

# Add a larger arrow to indicate the counterclockwise direction
arrow_start_idx = 75  # Index for the starting point of the arrow
plt.annotate(
    "",  # Empty text for arrow
    xy=(x[arrow_start_idx - 1], y[arrow_start_idx - 1]),  # Arrow end point
    xytext=(x[arrow_start_idx], y[arrow_start_idx]),      # Arrow start point
    arrowprops=dict(arrowstyle="->", lw=2, color="red", mutation_scale=20),
)

# Plot two points and label them as "State 1" and "State 2"
state1_idx = 10
state2_idx = 60
plt.plot(x[state1_idx], y[state1_idx], 'bo')  # Point for State 1
plt.plot(x[state2_idx], y[state2_idx], 'go')  # Point for State 2
plt.text(x[state1_idx] + 0.1, y[state1_idx] + 0.1, 'State 1', color='blue')
plt.text(x[state2_idx] + 0.1, y[state2_idx] + 0.1, 'State 2', color='green')

# Make the plot square to show the circle accurately
plt.axis("equal")

# Hide x and y axis values
plt.xticks([])
plt.yticks([])

# Set limits to ensure the circle is fully in the top-right quadrant with buffer
plt.xlim(0, 2 * (r + buffer))
plt.ylim(0, 2 * (r + buffer))

# Labels and title
plt.xlabel("V")
plt.ylabel("P")
plt.title("Cyclic Process")

# Tight layout
plt.tight_layout()

# Save the plot
plt.savefig("./images/cyclic_process.png")

```
